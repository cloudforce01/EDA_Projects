```python
# Importing necessary packages for data analysis.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.ticker as mtick
import seaborn as sns
import warnings
from typing import Optional
from IPython.display import display

# Suppress FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

# The below dictionary sets the common properties of any type of chart and this would standardize the properites for all type of charts.
SET_CHART_PROPERTIES: dict = {
                               'fontsize': 10.0, 'fontweight': 'normal', 'rotation': 360,'ha': 'center', 'va': 'bottom',
                               'width': 10.0, 'length': 5.0, 'bar_width': 1.0, 'autopct': '%1.2f%%', 'color': 'darkmagenta',
                               'percent_format': False, 'encoding': 'latin-1','title_fontsize': 15.0, 'title_bgcolor': 'darkviolet', 
                               'title_color': 'white', 'title_fontweight': 'bold', 'chart_title': 'Graph/Chart',
                               'loc': 'upper right', 'bbox_to_anchor': (1.0, 1.0), 'bin_bar': 0.25,
                               'palette': None, 'hue': None
    
                             }

# Define bins ranges.
bin_ranges: dict = {
    'Not rated': [0.0, 0.0],  'Poor': [1.8, 2.4],  'Average': [2.5, 3.4], 
    'Good': [3.5, 3.9], 'Very Good': [4.0, 4.4], 'Excellent': [4.5, 4.9]
}

# The below dictionary sets the error messages.
DICT_ERROR_MESSAGE: dict[str, str] = {
    'invalid_arguments_err_msg': 'Check the arguments passed to the function and try again.'
}

# 1. Data Ingestion:

def load_data(file_path: str, encoding: str = SET_CHART_PROPERTIES['encoding']) -> Optional[pd.DataFrame]:
    '''
    Load data from a CSV file.

    Parameters:
    file_path (str): Path to the CSV file.
    encoding (str): File encoding.
    
    Returns:
    pd.DataFrame or None: Loaded DataFrame or None if an error occurs.
    '''
    try:
        if file_path.split('.')[-1].upper() == 'CSV':
            df: pd.DataFrame = pd.read_csv(file_path, encoding=encoding)
        elif 'XLS' in file_path.split('.')[-1].upper():
            df: pd.DataFrame = pd.read_excel(file_path)
        else:
           print(f'Supported file types are csv for zomato data and MS Excel for Country Code data.')
           return None
        return df
    except FileNotFoundError:
        print(f"Error: File not found at path: {file_path}")
        return None
    except Exception as e:
        print(f"Error loading data: {e}")
        return None
        
def merge_dataframes(zomato_data, country_code_data) -> Optional[pd.DataFrame]:
    '''
    To merge two dataframes.

    Parameters:
    zomato_data (str): .csv file path.
    country_code_data (str): .xls or .xlsx - MS Excel file path.

    Returns:
    pd.DataFrame or None - if arguments passed are invalid
    '''
    try:
        return pd.merge(zomato_data, country_code_data, on='Country Code', how='left')
    except Exception as e:
        print(f'Error merging dataframes: {e}')
        return None
        
# Source files

zomato_file_path: str = 'Zomatodataset\zomato.csv'
country_code_file_path: str = 'Zomatodataset\Country-Code.xlsx'

# loading source files into respective dataframes.
df_zomato: pd.DataFrame = load_data(zomato_file_path) 
if df_zomato is not None:
    df_country: pd.DataFrame = load_data(country_code_file_path) 
if df_zomato is not None and df_country is not None:
    final_df = merge_dataframes(df_zomato, df_country)
else:
    exit()
    
# 2. Exploratory Data Analysis (EDA):

def display_eda(df: pd.DataFrame, rows: int=5):
    """
    Perform exploratory data analysis on a DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.
    rows (int): Number of rows to display for head and tail.
    """
    if isinstance(df, pd.DataFrame) and bool(df.empty) == False:
        print('\nData Exploration:\n')
        print(df.info(), end='\n\n')  
        display(df.describe())
        print(f"The shape of the dataframe is: {df.shape}", end ='\n\n')
        print('Columns:', df.columns.tolist())
    else:
        print('Data is not loaded properly into the dataframe and hence could not display exploratory data analysis results. Check the file path and file name while loading the data')


def report_missing_values(df: pd.DataFrame, print_null_cols: bool = False) -> Optional[pd.DataFrame]:
    """
    Display missing values for each column in a DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.

    Returns:
    pd.DataFrame or None - if arguments passed are invalid.
    """
    if isinstance(df, pd.DataFrame):
        null_summary: pd.DataFrame = df.isnull().sum()
        missing_cols: list[tuple[str, int]] = [(column, count) for column, count in null_summary.items() if count > 0]
        null_summary = null_summary.reset_index().rename(columns = {0: 'null_count'})
        if print_null_cols == True:
            print(f"\nColumns with missing values: {missing_cols}")
        return null_summary
    else:
        print('Data is not loaded properly into the dataframe and hence could not display missing values information. Check the file path and file name while loading the data')
        return None
        
# 3. Visualization:


def input_validation(lst_flt_no: Optional[list] = [], lst_int_no: Optional[list] = [], \
                          lst_txt: Optional[list] = [], lst_bool: Optional[list] = [], lst_df: Optional[list] = [], \
                          lst_plt: Optional[list] = []) -> bool:
    '''
    Checks the parameter(s) type passed to function is valid or not.
    
    Parameters:
    lst_flt_no (list[float]]: To check if the parameters are float in one go by using list.
    lst_int_no (list[int]]: To check if the parameters are int in one go by using list.
    lst_txt (list[str]]: To check if the parameters are str in one go by using list.
    lst_bool (list[bool]]: To check if the parameters are bool in one go by using list.
    lst_df (list[pd.DataFrame]]: To check if the parameters are pd.DataFrame in one go by using list.
    lst_plt (list[plt]]: To check if the parameters are not None in one go by using list.

    Returns:
    bool or None type.
    
    '''
    
    valid_float_type = all([isinstance(element, (int, float)) for element in lst_flt_no]) if len(lst_flt_no) > 0  else True
    valid_int_type = all([isinstance(element, int) for element in lst_int_no]) if len(lst_int_no) > 0 else True
    valid_str_type =  all([isinstance(element, (str, pd.Series)) for element in lst_txt]) if  len(lst_txt) > 0  else True  
    valid_bool = all([isinstance(element, bool) for element in lst_bool]) if len(lst_bool) > 0  else True
    # valid_dataframe = all([isinstance(element, pd.DataFrame) for element in lst_df if all(bool(element).empty) == False]) if len(lst_df) > 0 else True
    return valid_float_type and valid_int_type and valid_str_type and valid_bool 
    # and valid_dataframe 

def custom_palette(df: pd.DataFrame, column: str):
    '''
    This is to define and store custom palette

    Parameters:
    df (pd.DataFrame): This is a dataframe.
    column (str): Dataframe column

    Returns:
    List of colors without space inbetween.
    '''
    p: list = [color.replace(' ', '') for color in df[column].unique().tolist()]
    return p


def set_chart_title(plt,  ctitle: str, fontsize: float, backgroundcolor: str, color: str, fontweight: str) -> None:
    '''
    It sets chart title for the graph with fontsize, fontwieght, color, backgroundcolor.

    Parameters:
    plt: Referencing or Namespace for matplotlib.pyplot
    ctitle (str):  Title of the chart
    fontsize (float): Fontsize of the title
    backgroundcolor (str): Background color of thet title
    color (str): Color of the chart title's text.
    fontweight (str): Fontweight of chart's title.


    Returns:
    None
    '''
    
    if input_validation(lst_plt = [plt], lst_flt_no = [fontsize], lst_txt = [backgroundcolor, color, fontweight, ctitle]) == True:
        
        try:
            plt.title(ctitle).set_fontsize(fontsize)    
            plt.title(ctitle).set_backgroundcolor(backgroundcolor)
            plt.title(ctitle).set_color(color)
            plt.title(ctitle).set_fontweight(fontweight)   
        except Exception as e:   
            print(f"{DICT_ERROR_MESSAGE['invalid_arguments_err_msg']}")
    else:
         print(f"{DICT_ERROR_MESSAGE['invalid_arguments_err_msg']}")

def plot_bar(x: str, y: str, data: Optional[pd.DataFrame], width: float = SET_CHART_PROPERTIES['width'], \
            length: float = SET_CHART_PROPERTIES['length'], palette: list = SET_CHART_PROPERTIES['palette'], hue: list = SET_CHART_PROPERTIES['hue'], \
            rotation: int = SET_CHART_PROPERTIES['rotation'], fontsize: float = SET_CHART_PROPERTIES['fontsize'],\
            fontweight: str = SET_CHART_PROPERTIES['fontweight'], color: str = SET_CHART_PROPERTIES['color'], \
            bar_width: float = SET_CHART_PROPERTIES['bar_width'], percent_format: bool = SET_CHART_PROPERTIES['percent_format'],\
            chart_title: str = SET_CHART_PROPERTIES['chart_title'], title_bgcolor: str = SET_CHART_PROPERTIES['title_bgcolor'], \
            title_color: str = SET_CHART_PROPERTIES['title_color'], title_fontweight: str = SET_CHART_PROPERTIES['title_fontweight'],\
            title_fontsize: float = SET_CHART_PROPERTIES['title_fontsize'], loc: str = SET_CHART_PROPERTIES['loc'],\
            bbox_to_anchor: tuple = SET_CHART_PROPERTIES['bbox_to_anchor'], ha: str = SET_CHART_PROPERTIES['ha'],\
            va: str = SET_CHART_PROPERTIES['va']):

    """
    Plot a bar chart.

    Parameters:
    x (str): Column name for x-axis.
    y (str): Column name for y-axis.
    data (pd.DataFrame): Data to plot.    
    width (float): Width of the plot.
    length (float): Length of the plot.        
    palette (list): List of color palette.
    hue (str): column of the dataframe and this column should contain color.
    rotation (int): Rotation angle for x-axis labels.        
    fontsize (float): fontsize of x and y labels.
    fontweight (str): fontweight of x and y labels.        
    color (list): list of color for each bar. 
    bar_width (int): Width of each bar.   
    percent_format (bool): formatting the percentage based on this value.
    chart_title (str): title of the chart.
    title_bgcolor (str): Background color of the chart title.
    title_color (str): Font color of the tite.
    title_fontweight (str): Font weight of the title.
    title_fontsize (float): font size of the title.
    loc (str): Location of the legend box in the chart.
    bbox_to_anchor (tuple(float, float)): Pixel position of the legend box.
    ha (str): Alignment of Horizontal axis.
    va (str): Alignment of vertical axis.
    
    """
   
    if input_validation(lst_flt_no = [width, length, fontsize, bar_width, title_fontsize], lst_int_no = [rotation],\
                        lst_txt = [x, y, fontweight, color, chart_title, title_bgcolor, title_color, title_fontweight, loc, ha, va],\
                        lst_bool = [percent_format], lst_df = [data]) == True:
        try:
            
            fig, ax = plt.subplots(figsize=(width, length))
            bar_plot: plt.AxesSubplot = sns.barplot(x = x, y = y, data = data, width = bar_width, hue = hue, palette = palette)
            ax: plt.Axes = plt.gca()
            plt.xticks(rotation = rotation, fontsize = fontsize,\
                           fontweight = fontweight)
            ax.set_xlabel(x)
            
            for i, bar in enumerate(bar_plot.patches):
                x: int = bar.get_x() + bar.get_width() / 2
                y: int = bar.get_height()
                if percent_format == False:
                    ax.text(x, y, f'{int(y) if not np.isnan(y) else ""}', ha = ha, va = va, \
                            fontsize = fontsize)
                else:
                    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
                    ax.text(x, y, '{:.1f}%'.format(y), ha = ha, va = va, \
                            fontsize = fontsize)
                    
            set_plt_display(plt = plt, chart_title = chart_title, title_fontsize = title_fontsize, \
                            title_fontweight = title_fontweight, title_bgcolor = title_bgcolor,\
                             title_color = title_color, width = width, length = length, hue = hue, loc = loc, bbox_to_anchor = bbox_to_anchor)
        except Exception as e:
            print(f'Might be matplotlib.pyplot or plt is invalid. Check the arguements passed to the function and then try again: {e}')
        
    else:
        print(DICT_ERROR_MESSAGE['invalid_arguments_err_msg'])


def set_plt_display(plt, chart_title: str, title_fontsize: float , title_fontweight: str, title_bgcolor: str, \
                    title_color: str, width: float, length: float, hue: list, loc: str, bbox_to_anchor: tuple[float, float]):   
    '''
    To display the plot with customized settings.

    Parameters:
    plt: Plot
    chart_title (str): Chart title.
    title_fontsize (float): Font size of the title.
    title_fontweight (str): Font weight of the title.
    title_bgcolor (str): Backgrond color of the title.
    title_color (str): Font color of the title.
    width (float): Width of the plot.
    length (float): Length of the plot.
    hue (list): To display the legend.
    '''
    
    # Formatting.
    
    set_chart_title(plt, ctitle = chart_title, fontsize = title_fontsize, fontweight = title_fontweight, backgroundcolor = title_bgcolor,\
                   color = title_color)
    plt.tight_layout()    
    if hue is not None:
        plt.legend(loc = loc, bbox_to_anchor = bbox_to_anchor)
    plt.show()

def plot_pie(x: Optional[str], labels: Optional[str], data: Optional[pd.DataFrame], \
             autopct: str = SET_CHART_PROPERTIES['autopct'], color: str = SET_CHART_PROPERTIES['color'], \
             fontsize: float = SET_CHART_PROPERTIES['fontsize'], fontweight: str = SET_CHART_PROPERTIES['fontweight'],\
             width: float = SET_CHART_PROPERTIES['width'], length: float = SET_CHART_PROPERTIES['length'],
             title_fontweight: str = SET_CHART_PROPERTIES['title_fontweight'], title_fontsize: float = SET_CHART_PROPERTIES['title_fontsize'],
             title_bgcolor: str = SET_CHART_PROPERTIES['title_bgcolor'], title_color: str = SET_CHART_PROPERTIES['title_color'],
             chart_title: str = SET_CHART_PROPERTIES['chart_title']):
    """
    Plot a pie chart.
    
    Parameters:
    x (str): Column name 
    labels (str): Column name for labels.
    data (pd.DataFrame): Data to plot. 
    autopct (str): Format the float value.
    color (str): Font color.
    fontsize (int): Text font size.
    fontweight (str): Text fontweight like normal, bold etc.
    width (float): Width of the plot.
    length (float): Length of the plot.
    title_fontweight (str): Font weight of the chart title.
    title_fontsize (float): Font size of the title.
    title_bgcolor (str): Background color of the title.
    title_color (str): Font color of the title.
    chart_title (str): Title of the chart.
    
    """
    if input_validation(lst_flt_no = [width, length, fontsize, title_fontsize],\
                        lst_txt = [x, labels, fontweight, color, chart_title, title_color, title_bgcolor, title_fontweight, autopct], \
                        lst_df = [data]) == True:
        try:
            # plt.rcParams['figure.figsize'] = (width, length)
            fig, ax = plt.subplots(figsize=(width, length))
            plt.pie(x=x, labels = labels, autopct = autopct, textprops={'fontsize': fontsize, 'fontweight': fontweight}) 
            set_plt_display(plt = plt, chart_title = chart_title, title_fontsize = title_fontsize, \
                            title_fontweight = title_fontweight, title_bgcolor = title_bgcolor,\
                             title_color = title_color, width = width, length = length, hue = None, loc = None, bbox_to_anchor = None)
        except Exception as e:
             print(f'Might be matplotlib.pyplot or plt is invalid. Check the arguements passed to the function and then try again: {e}')
    else:
        print('Data is not loaded properly into the dataframe and hence pie chart is not getting displayed. Check the file path and file name')


# Reverse mapping of bin_ranges for bin assignment
# bin_labels: dict = {v[0]: k for k, v in bin_ranges.items()}

# Assign bins to rows
def assign_bin(rating: pd.Series):
    '''
    To assign bins

    Parameters:
    rating (pd.Series): Data consists in a dataframe

    Returns:
    str or None - If there is an error in data
    '''
    
    for label, (low, high) in bin_ranges.items():
        if low <= rating <= high:
            return f"{low}-{high}" if label != 'Not rated' else f"{low}"
    return None

customer_ratings: pd.DataFrame = final_df.groupby(['Aggregate rating', 'Rating text', 'Rating color']).size().\
                            reset_index().rename(columns = {0: 'RatingCount'})
customer_ratings['Rating bin'] = customer_ratings['Aggregate rating'].apply(assign_bin)

def agg_bins(data: pd.DataFrame, ha: str = SET_CHART_PROPERTIES['ha'], va: str = SET_CHART_PROPERTIES['va'],\
                rotation: int = SET_CHART_PROPERTIES['rotation'], fontsize: float = SET_CHART_PROPERTIES['fontsize'],\
                fontweight: str = SET_CHART_PROPERTIES['fontweight'], width: float = SET_CHART_PROPERTIES['width'],\
                length: float = SET_CHART_PROPERTIES['length'], chart_title: str = SET_CHART_PROPERTIES['chart_title'],\
                title_fontsize: float = SET_CHART_PROPERTIES['title_fontsize'], title_fontweight: str = SET_CHART_PROPERTIES['title_fontweight'],\
                title_bgcolor: str = SET_CHART_PROPERTIES['title_bgcolor'], title_color: str = SET_CHART_PROPERTIES['title_color'],\
                loc: str = SET_CHART_PROPERTIES['loc'], bbox_to_anchor: tuple = SET_CHART_PROPERTIES['bbox_to_anchor'],\
                hue: list = None, bin_bar: float = SET_CHART_PROPERTIES['bin_bar'], color: str = SET_CHART_PROPERTIES['color']):
    '''
    Aggregate data by bins

    Parameters:
    ratings (pd.DataFrame): Contains data.
    ha (str): sets alignment of horizontal axis.
    va (str): sets alignment of vertical axis.
    rotation (int): sets x-axis rotation.
    fontsize (float): Text fontsize.
    fontweight (str): Text fontweight.
    width (float): Width of chart.
    length (float): Length of chart.
    chart_title (str): Title of the chart.
    title_fontsize (float): Font size of the chart title.
    title_fontweight (str): Font weight of the title.
    title_bgcolor (str): Background color of the title.
    title_color (str): Font color of the title.
    loc (str): Location of the legend in the chart.
    bbox_to_anchor (tuple(float, float)): pixel position of the legend box.
    hue (list(str)): Displays the legend.
    bin_bar (float): Width of the individual bar.
    color (str): Font color
    '''
    if input_validation(lst_flt_no = [width, length, fontsize, title_fontsize, bin_bar],\
                        lst_txt = [fontweight, color, chart_title, title_color, title_bgcolor, title_fontweight, ha, va], \
                        lst_int_no = [rotation], lst_df = [data]) == True:
        agg_df: pd.DataFrame = data.groupby('Rating bin').agg({
            'RatingCount': 'sum',
            'Rating text': 'first'
        }).reset_index()
        p = custom_palette(ratings, 'Rating color')
        # Plot
       
        fig, ax = plt.subplots(figsize=(width, length))
        bars = ax.bar(agg_df['Rating bin'], agg_df['RatingCount'], color =  p, width = bin_bar )
        
        # Add Rating text on the bars
        for bar, text in zip(bars, agg_df['Rating text']):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width() / 2, height - height * 0.1, text, ha = ha, va = va, \
                    rotation = rotation, fontsize = fontsize,\
                    fontweight = fontweight)
        set_plt_display(plt = plt, chart_title = chart_title, title_fontsize = title_fontsize, \
                                title_fontweight = title_fontweight, title_bgcolor = title_bgcolor,\
                                 title_color = title_color, width = width, length = length, loc = loc, bbox_to_anchor = bbox_to_anchor, hue = hue)
    else:
        print(DICT_ERROR_MESSAGE['invalid_arguments_err_msg'])
```
